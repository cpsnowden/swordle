# EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def eda_plots(df, set_lims=True):
    
    centroids = pd.DataFrame({
        "X": df.filter(regex = "_X$").mean(axis = 1),
        "Y": df.filter(regex = "_Y$").mean(axis = 1),
        "Z": df.filter(regex = "_Z$").mean(axis = 1),
        "TARGET": df["TARGET"]
    })
    
    print(df.shape)
    
    plt.figure(figsize=(20,5))
    ax = sns.countplot(data=df, x = "TARGET");
    ax.set_title("Target Distributions")
    plt.show()
    
    plt.figure(figsize=(12,10))
    ax = sns.kdeplot(data=centroids, x="X", y="Y", fill=True, cbar=True)
    if set_lims:
        ax.set_xlim(0,1)
        ax.set_ylim(0,1)
    ax.set_title("Centroid Distribution")
    ax.invert_yaxis();
    
    plt.show()
    g = sns.FacetGrid(centroids, col="TARGET", col_wrap=6, sharex=True, sharey=True, xlim=(0,1), ylim=(1, 0))
    g.figure.suptitle("Centroid Distribution", size=16)
    g.map(sns.kdeplot, "X", "Y", fill=True, cbar=True)
    for ax in g.axes:
        ax.invert_yaxis()
```
